```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://in.indeed.com/jobs?q=data+scientist&l=India")
page
```

```python
page.content
```

```python
# Now we will parse this source code by using Beautiful Soup and the parser used here is html.parser as the source code is in HTML
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
#First we will extract the html tag where we have the first job title
```

```python
first_title=soup.find('div',class_="jobsearch-SerpJobCard-footer")
first_title
```

```python
#Print first job title
first_title.text
```

```python
first_title.text.replace('\n\n\n','')
```

```python
first_company=soup.find('div',class_='icl-u-lg-mr--sm icl-u-xs-mr--xs').get_text()
first_company
```
